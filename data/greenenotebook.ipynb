```python
total_countries =helth_df.groupby(['Indicator Name']).sum(["1960"]).sort_values(["1960"],ascending=False)
total_countries = total_countries.T.iloc[15:]
total_countries.T.head(300)


```

```python
#total_countries.plot.scatter(x=total_countries['Literacy rate, adult male (% of males ages 15 and above)'],
#                             y=total_countries[14])


# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(total_countries['Literacy rate, adult male (% of males ages 15 and above)'], total_countries['% of males ages 15-49 having comprehensive correct knowledge about HIV (2 prevent ways and reject 3 misconceptions)'],s=0.00000001*total_countries['Population, male'])
# set a title and labels
ax.set_title('How much do men know HIV')
ax.set_xlabel('Literacy rate in males')
ax.set_ylabel('knowledge about HIV in males')
```

```python
sn.lineplot(data=total_countries['Literacy rate, adult male (% of males ages 15 and above)'])
```

```python
sn.lineplot(data=total_countries['% of males ages 15-49 having comprehensive correct knowledge about HIV (2 prevent ways and reject 3 misconceptions)'])
```

```python
only_hp_params = total_countries[['% of males ages 15-49 having comprehensive correct knowledge about HIV (2 prevent ways and reject 3 misconceptions)',
                                  'Literacy rate, adult male (% of males ages 15 and above)','Urban population','Rural population']] 
only_hp_params
```

```python
corr_matrix= only_hp_params.corr(min_periods=30)

## Heatmap showing the highest/lowest rate of each of the five crimes
plt.figure(figsize=(12, 9))
sn.heatmap(corr_matrix, annot=True,vmin= -1, vmax=1, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
