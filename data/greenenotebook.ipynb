```python
#Loading in the desired libraries to play with

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
#Loading in the desired dataframe

helth_df = pd.read_csv("data.csv",index_col=0)
helth_df
```

```python
#Listing out columns to get a better grasp on the amount of data we're dealing with

helth_df.columns
```

```python
#Grouping by indicator agnostic of country and summing all other columns together to get data for the whole world for each indicator

total_countries =helth_df.groupby(['Indicator Name']).sum(["1960"]).sort_values(["1960"],ascending=False)

#Flipping the axes to sort by indicator each year
total_countries = total_countries.T.iloc[15:]
'''
Also retroactively came back and parsed out all data before the 15th row; 
After having chosen the parameters for this analysis, the data suggested 
that worlwide literacy was in question before 1975 i.e. somehow we put
a man on the moon despite no one on the planet being literate... started
looking only at data following consistent literacy records as a result, 
which started at 1975!
'''
#temporarily flipping back just to get a look at the data from this perspective
total_countries.T.head(300)


```

```python
'''
Parsed down the data further to only include parameters we were looking to find a correlation in (and passed it into a new variable): 
'% of males ages 15-49 having comprehensive correct knowledge about HIV (2 prevent ways and reject 3 misconceptions)',
'Literacy rate, adult male (% of males ages 15 and above)',
'Urban population',
and 'Rural population'.
Additionally took this opportunity to cut out the unnamed index row.
'''

only_hp_params = total_countries[['% of males ages 15-49 having comprehensive correct knowledge about HIV (2 prevent ways and reject 3 misconceptions)',
                                  'Literacy rate, adult male (% of males ages 15 and above)','Urban population','Rural population']][:-1] 
only_hp_params.rename(columns={'% of males ages 15-49 having comprehensive correct knowledge about HIV (2 prevent ways and reject 3 misconceptions)': 'Males w/ HIV knowledge','Literacy rate, adult male (% of males ages 15 and above)': 'Male literacy rate'}, inplace=True)
#renamed the indicators to be a bit more concise... Be sure to speak to this while presenting! the data is the same!
```

```python
corr_matrix= only_hp_params.corr(min_periods=35)

## Heatmap showing 
plt.figure(figsize=(12, 9))
sn.heatmap(corr_matrix, annot=True,vmin= 0, vmax=1, cmap='coolwarm')
plt.title('Correlation Heatmap',size=30)
plt.xlabel('Indicator Name', size = 20)
plt.ylabel('Indicator Name', size = 20)
plt.show()

'''
Created the heatmap here. known issues with annotation persist.
Additionally elected to start cmap min value at 0 soas not to confuse despite 
most data being aggregated above 0.4. because of that the gradient isn't as well captured
but not starting at 0 I felt would be too misleading.
'''
```
